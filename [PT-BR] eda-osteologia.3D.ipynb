### **Visualização dos estudantes do período 2024.2 (1° período)**

```python
df_2024_2 = df[df['semestre'] == '2024.2 (estou cursando)']

contagem_confiança = df_2024_2['confiança_osteologia'].value_counts().sort_index()

contagem_confiança_percentual = (contagem_confiança / contagem_confiança.sum()) * 100

plt.figure(figsize=(8, 5))
ax = sns.barplot(x=contagem_confiança.index, y=contagem_confiança.values, color='#612d82')

max_val = contagem_confiança.values.max()
ax.set_ylim([0, max_val + 2])

for i, valor in enumerate(contagem_confiança_percentual):
    ax.text(i, contagem_confiança.values[i] + 0.5, f'{valor:.1f}%',
            ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')

plt.xlabel('Nível de Confiança')
plt.ylabel('Número de Estudantes')
plt.title('Nível de confiança em osteologia - Estudantes do período 2024.2', pad=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

É importante destacar que os estudantes do **1° período (2024.2)**, quando comparados aos demais, tiveram um contato mais recente com o aplicativo 3D, já que **estavam cursando a disciplina de Anatomia Descritiva no momento da aplicação da pesquisa**.

Percebemos que:

  - Pouco mais da metade dos estudantes relataram sentir-se parcialmente confiantes em relação à osteologia.
  - Dois estudantes afirmaram não se sentir confiantes no conhecimento em osteologia, mesmo após utilizarem a plataforma web 3D.
